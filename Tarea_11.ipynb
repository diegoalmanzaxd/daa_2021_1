```
class NodoArbol:
  def __init__(self,value,left=None,rigth=None):
    self.data=value
    self.left=left
    self.rigth=rigth
    self.ban=0
    self.ban1=0
```

```
class BinarySearchTree:
  def __init__(self):
    self.__root=None
  def insert(self,value):
    if self.__root==None:
      self.__root=NodoArbol(value,None,None)
    else:
      self.__insert_nodo__(self.__root,value)
  
  def __insert_nodo__(self,nodo,value):
    if nodo.data==value:
      pass
    if value<nodo.data:#si es verdadero va a la IZQ
      if nodo.left==None:#si hay espacio en la izq, ahi va
        nodo.left=NodoArbol(value,None,None)#insertamos el nodo
      else:
        self.__insert_nodo__(nodo.left,value) #buscar el sub arbol izq
    elif value>nodo.data:
      if nodo.rigth==None:
        nodo.rigth=NodoArbol(value,None,None)
      else:
        self.__insert_nodo__(nodo.rigth,value)
  def buscar(self,value):
    if self.__root==None:
      return False
    else:
      return self.__busca_nodo(self.__root,value)
  def __busca_nodo(self,nodo,value):
    if nodo==None:
      return None
    elif nodo.data==value:
      return nodo.data
    elif value<nodo.data:
      return self.__busca_nodo(nodo.left,value)
    else:
      return self.__busca_nodo(nodo.rigth,value)
  def transversal(self,format="inorden"):
    if format=="inorden":
      self.__recorrido_in(self.__root)
    elif format=="preorden":
      self.__recorrido_pre(self.__root)
    elif format=="posorden":
      self.__recorrido_pos(self.__root)
    else:
      print("Formato de corrido no valido")
  def __recorrido_pre(self,nodo):
    if nodo!=None:
      print(nodo.data)
      self.__recorrido_pre(nodo.left)
      self.__recorrido_pre(nodo.rigth)
  def __recorrido_in(self,nodo):
    if nodo!=None:
      self.__recorrido_in(nodo.left)
      print(nodo.data)
      self.__recorrido_in(nodo.rigth)
  def __recorrido_pos(self,nodo):
    if nodo!=None:
      self.__recorrido_pos(nodo.left)
      self.__recorrido_pos(nodo.rigth)
      print(nodo.data)

  def eliminar(self, value):
    mov=[]
    bandera1=0
    bandera2=0
    bandera3=0
    aux=self.__root
    x=self.__root
    while bandera1==0 and bandera3==0:
      aux=self.__root
      bandera2=0

      for i in mov:
        if i=="L":
          aux=aux.left
        if i=="R":
          aux=aux.rigth

      if len(mov)==0 and self.__root.left==None and self.__root.rigth==None:
        bandera2=1
        if self.__root.data ==value:
          bandera3=1
        else:
          bandera1=1
      if self.__root.left==None and self.__root.rigth!=None:
        if len(mov)==0 and self.__root.rigth.ban==1:
          bandera1==0
      if self.__root.left!=None and self.__root.rigth==None:
        if len(mov)==0 and self.__root.rigth.ban==1:
          bandera1==0
      if self.__root.left!=None and self.__root.rigth!=None:
        if len(mov)==0 and self.__root.left.ban==1 and self.__root.rigth.ban==1:
          bandera1=1
      if aux.data==value and bandera3==0:
        for i in range(len(mov)-1):
          if mov[i]=="L":
            x=x.left
          if mov[i]=="R":
            x=x.rigth
        bandera3=1
        bandera2=1
        
      if aux.left!= None and aux.left.ban==0 and bandera2==0:
        aux=aux.left
        aux.ban=1
        mov.append("L")
        bandera2=1
      if aux.rigth!=None and aux.rigth.ban==0 and bandera2==0:
        aux=aux.rigth
        aux.ban=1
        mov.append("R")
        bandera2=1
      if aux.rigth==None and aux.left==None and bandera2==0:
        mov.pop()
      if aux.left!=None and aux.rigth!=None and aux!=self.__root and bandera2==0:
        if aux.left.ban==1 and aux.rigth.ban==1:
          mov.pop()
      if aux.rigth!=None and bandera2==0:        
        if aux.left==None and aux.rigth.ban==1:
          mov.pop()
      if aux.left!=None and bandera2==0:
        if aux.left.ban==1 and aux.rigth==None:
          mov.pop()
    if bandera1==1:
      print("No existe el valor que deseas borrar")
    elif bandera3==1:
      ban1=0
      if len(mov)==0:
        if x.left==None and x.rigth==None:
          x.data=None
          ban1=1
        if x.left!=None and x.rigth!=None:
          ban1=2
      elif len(mov)!=0:
        if mov[len(mov)-1] =="R":
          if x.rigth.left!=None and x.rigth.rigth!=None:
            ban1=3
        if mov[len(mov)-1] =="L" and x.left!=None :
          if x.left.left!=None and x.left.rigth!=None:
            ban1=4    

      bandera=0
      mov2=[]
      lis1=[]
      lis2=[]
      aux=None
      if ban1==2 or ban1==3 or ban1==4:
        while bandera==0:
          bandera2=0
          if ban1==2:
            aux=x.rigth
          if ban1==3:
            aux=x.rigth.rigth
          if ban1==4:
            aux=x.left.rigth
          if aux.left==None and aux.rigth==None:
            lis1.append(aux.data)
            if ban1==2:
              aux=x
              print(aux.data)
            if ban1==3:
              aux=x.rigth
            if ban1==4:
              aux=x.left
            aux.data=lis1[0]
            aux.rigth=None
            bandera=1
            bandera2=1
          if aux.left!=None and aux.rigth!=None:
            if len(mov2)==0 and aux.left.ban1==1 and aux.rigth.ban1==1:
              bandera=1
          if aux.left==None and aux.rigth!=None:
            if len(mov2)==0 and aux.rigth.ban1==1:
              bandera=1
          if aux.left!=None and aux.rigth==None:
            if len(mov2)==0 and aux.left.ban1==1:
              bandera=1
          for i in mov2:
            if i=="L":
                aux=aux.left
            if i=="R":
                aux=aux.rigth
          if aux.left!= None and aux.left.ban1==0 and bandera2==0:
            aux=aux.left
            aux.ban1=1
            mov2.append("L")
            bandera2=1
          if aux.rigth!=None and aux.rigth.ban1==0 and bandera2==0:
            aux=aux.rigth
            aux.ban1=1
            mov2.append("R")
            bandera2=1
          if aux.rigth==None and aux.left==None:
            movimientos=[]
            for i in mov2:
              movimientos.append(i)
            lis1.append(aux.data)
            lis2.append(movimientos)
            mov2.pop()
          if ban1==2:
            arb=x.rigth
          if ban1==3:
            arb=x.rigth.rigth
          if ban1==4:
            arb=x.left.rigth
          if aux.left!=None and aux.rigth!=None and aux!=arb:
            if aux.left.ban1==1 and aux.rigth.ban1==1:
                mov2.pop()
          if aux.rigth!=None:        
              if aux.left==None and aux.rigth.ban1==1:
                mov2.pop()
          if aux.left!=None and aux!=arb:
              if aux.left.ban1==1 and aux.rigth==None:
                mov2.pop()
      
        aux32=[]
        for i in lis1:
          aux32.append(i)
        aux32.sort()
        cont=0
        for i in range(len(aux32)):
          if lis1[i]==aux32[0]:
            cont=i
        if len(lis2)!=0:
          if len(lis2[cont])==1:
            if lis2[cont][0]=="L":
              aux.left=None
            if lis2[cont][0]=="R":
              aux.rigth=None
          if len(lis2[cont])!=1 and lis2[cont]!=0:
            for i in range(len(lis2[cont])-1):
              if lis2[cont][i]=="L":
                aux=aux.left
              if lis2[cont][i]=="R":
                aux=aux.rigth
            if lis2[cont][len(lis2[cont])-1]=="L":
              aux.left=None
            if lis2[cont][len(lis2[cont])-1]=="R":
              aux.rigth=None
        if ban1==2:
          aux=x
        if ban1==3:
          aux=x.rigth
        if ban1==4:
          aux=x.left
        aux.data=lis1[cont]
      
      ban=0
      if ban1==0:
        if mov[len(mov)-1] =="R":
          if x.rigth.left!=None and x.rigth.rigth==None and ban==0:
            x.rigth=x.rigth.left
            ban=1
          if x.rigth.left==None and x.rigth.rigth!=None and ban==0:
            x.rigth=x.rigth.rigth
            ban=1
          if x.rigth.left==None and x.rigth.rigth==None and ban==0:
            x.rigth=None
        if mov[len(mov)-1] =="L":
          if x.left.left!=None and x.left.rigth==None and ban==0:
            x.left=x.left.left
            ban=1
          if x.left.left==None and x.left.rigth!=None and ban==0:
            x.left=x.left.rigth
            ban=1
          if x.left.left==None and x.left.rigth==None and ban==0:
            x.left=None
```
